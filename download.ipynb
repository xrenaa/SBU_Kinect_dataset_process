```python
import os
import time
import zipfile
def unzip(sourceFile, targetPath):
    '''
    :param sourceFile: zip
    :param targetPath: 
    :return:
    '''
    file = zipfile.ZipFile(sourceFile, 'r')
    file.extractall(targetPath)
    print('success to unzip file!')

source="./dataset"
target="./unzip"
for i in range(1,22):   
    filename="demo_"+str(i)
    path=os.path.join(source,filename+".zip")
    print(path)
    target_path=os.path.join(target,str(i))
    unzip(path,target_path)
```

```python
import matplotlib.pyplot as plt
import argparse
import numpy as np
import torch
total=[]
labels=[]
for i in range(1,22): 
    filename='%02d' % i
    mainFile=os.path.join(target,filename)
    #print(mainFile)
    subFile=os.listdir(mainFile)[0]
    #print(subFile)
    sub_path=os.path.join(mainFile,subFile)
    cats = sorted(os.listdir(sub_path))[1::]
    #print(cat)
    for cat in cats:
        label=int(cat)-1
        #print(label)
        cat_path = os.path.join(sub_path,cat)
        nums = sorted(os.listdir(cat_path))
        if nums[0] == '.DS_Store':
            nums=nums[1::]
        for num in nums:
            list_1=[]
            list_2=[]
            one=[]
            num_path = os.path.join(cat_path,num)
            txt_path = os.path.join(num_path,"skeleton_pos.txt")
            #print(txt_path)
            with open(txt_path) as f:
                data = f.readlines()
            for row in data:
                posture = row
                pose_1=[]
                pose_2=[]
                posture_data = [x.strip() for x in posture.split(',')]

                joint_info = {}
                for i, n in enumerate(range(1, len(posture_data), 3)):
                    joint_info[i+1] = [(1280 - float(posture_data[n])*2560)/1280,
                                       (960 - (float(posture_data[n+1])/1.2)*1920)/960, posture_data[n+2]]

                person_1 = {k:joint_info[k] for k in range(1,16,1)}
                person_2 = {k-15:joint_info[k] for k in range(16,31,1)}
                #print(person_1)
    
                for key, value in person_1.items():
                    pose_1.append(value[0:2])
                array=np.array(pose_1)
                list_1.append(array)
    
                for key, value in person_2.items():
                    pose_2.append(value[0:2])
                array=np.array(pose_2)
                list_2.append(array)

            list_1=np.array(list_1)
            list_2=np.array(list_2)
            one.append(list_1)
            one.append(list_2)
            labels.append(label)
            total.append(one)          
```

# Part to interpolate the dataset all to length 25
### you can change the size=25 to the size that you want

```python
count=0
interpolate=[]
for i in range(282):
    slice=np.array(total[i])
    if slice.shape[1] > 35:
        count+=1
    temp=[]
    for person in range(2):
        pose=slice[person]
        tensor_1=torch.tensor(pose)
        tensor_1=tensor_1.permute(1,2,0)
        target_1=torch.nn.functional.interpolate(tensor_1, size=25, mode='linear')
        #print(target_1.shape)
        target_1=target_1.permute(2,0,1)
        temp.append(target_1.numpy())
    interpolate.append(np.array(temp))
interpolate=np.array(interpolate)
print(interpolate.shape)    
```

# Part to split the dataset to train_data and val_data 
## note that you should use the fuction below to save and load the json file

```python
import random
aa=[]
bb=[]
cc = list(zip(interpolate,labels))
random.shuffle(cc)
aa[:], bb[:] = zip(*cc)
print(len(aa), len(bb))
```

```python
import json
class NumpyEncoder(json.JSONEncoder):
    """ Special json encoder for numpy types """
    def default(self, obj):
        if isinstance(obj, (np.int_, np.intc, np.intp, np.int8,
            np.int16, np.int32, np.int64, np.uint8,
            np.uint16, np.uint32, np.uint64)):
            return int(obj)
        elif isinstance(obj, (np.float_, np.float16, np.float32, 
            np.float64)):
            return float(obj)
        elif isinstance(obj,(np.ndarray,)): #### This is the fix
            return obj.tolist()
        return json.JSONEncoder.default(self, obj) 
    
def save_to_json(dic,target_dir):
    dumped = json.dumps(dic, cls=NumpyEncoder)  
    file = open(target_dir, 'w')  
    json.dump(dumped, file)
    file.close()
    
def read_from_json(target_dir):
    f = open(target_dir,'r')
    data = json.load(f)
    data = json.loads(data)
    f.close()
    return data 
```

```python
train_dict={}
train_x=np.array(aa[0:198])
train_y=np.array(bb[0:198])
print(train_x.shape)
print(train_y.shape)
train_dict["x"]=train_x
train_dict["label"]=train_y
save_to_json(train_dict,"./json/train.json")
```
